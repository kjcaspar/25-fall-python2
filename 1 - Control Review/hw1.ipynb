```python
def is_perfect(input_int):
    """
    :param input_int: integer
    :return: Boolean stating whether num is 'perfect' meaning it is the sum of its factors
    """
    sum = 0
    for i in range(1,input_int):
        if input_int % i == 0:
            sum = sum + i
    if sum == input_int:
        return True
    else: 
        return False

print(is_perfect(100))
print(is_perfect(496))
```

```python
def lcm(int1, int2):
    """
    :param int1: integer
    :param int2: integer
    :return: the least common multiple of int1 and int2
    """
    multiple = 1
    while multiple%int1!= 0:
        multiple += 1
        while multiple%int2!= 0:
            multiple += 1
    return multiple

print(lcm(6,10))
print(lcm(3,30))
print(lcm(10,30))
```
